```python
print(len(test_data.targets))
preds = [5] * len(test_data.targets)
```

```python
def check_accuracy(preds, actual):
    if len(actual) != len(preds):
        return -1
    return sum([int(actual[i] == preds[i]) for i in range(0, len(actual))]) / len(actual)

print(check_accuracy([1, 1, 0], [0, 1, 0]))
```

```python
print(f"Accuracy: {check_accuracy(preds, test_data.targets)}")
```
